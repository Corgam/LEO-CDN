```python
import pandas as pd

df_cities = pd.read_csv("../../data/worldcities_all.csv")
```

```python
bbox = [44.6264, 0.7031, 57.7617, 20.8301]
min_lat, min_lng, max_lat, max_lng = bbox
```

```python
MARGIN_LAT = 2.5
MARGIN_LNG = 5
```

```python
df_bb = df_cities[
    (df_cities.lat > min_lat + MARGIN_LAT)
    & (df_cities.lat < max_lat - MARGIN_LAT)
    & (df_cities.lng > min_lng + MARGIN_LNG)
    & (df_cities.lng < max_lng - MARGIN_LNG)
]
```

```python
df_bb.shape
```

```python
NUM_GSTS = 50

df_bb_sample = df_bb.sample(n=NUM_GSTS, random_state=1)
```

```python
df_bb_sample
```

```python
df_bb_sample.to_csv("../../data/gsts_bb_sample.csv", index=False)
```
